```python
Cov_Data=r_data.cov()
mean_cov=Cov_Data.mean()
```

## qualified couples
### covariance > mean covariance
### correlation > 0.5

```python
qual_combi=np.zeros(Cov_Data.shape)
for i in range(len(data1['Symbol'].values)):
    for j in range(i):
        if np.array(Cov_Data)[i,j]<mean_cov[i] and np.array(Cov_Data)[i,j]<mean_cov[j] and np.array(Corr_Data)[i,j]<0.5:
            qual_combi[i,j]=1
```

```python
def add_candidates(couple, other_can):
    for other in other_can:
        count=0
        for cou in couple:
            if qual_combi[cou, other]==1 or qual_combi[other, cou]==1:
                count+=1
        if float(count)/len(couple)>=0.6:
            couple.append(other)
    return couple
```

## all portfolio 

```python
all_couples=[]
for c in range(len(data1['Symbol'].values)):
    couple=[]
    other_can=[]
    couple.append(c)
    for i in range(len(data1['Symbol'].values)):
        if i not in couple:
            other_can.append(i)
    couple=add_candidates(couple, other_can)
    couple=sorted(couple)
    all_couples.append(couple)

couples=[]
for i in all_couples:
    if i not in couples:
        couples.append(i)
couples
```

# Sharpe Ratio
### (Rp-Rf)/SIGMAp

```python
Rf=0.02
final_couples=[]
final_weights=[]
final_ratio=[]
final_rp=[]
for i in tqdm(couples):
    symbols=[]
    best_shar=0
    best_weight=np.zeros(len(i))
    best_rp=0
    for num in i:
        symbols.append(data1['Symbol'].values[num])
    mean_price=np.array(r_data[symbols].mean())*len(r_data)
    cov = np.array(r_data[symbols].cov()) 
    for count in range(1000):
        weights = [np.random.randint(1,100) for a in i]
        weights = np.array(weights, dtype=float)
        weights /= weights.sum()
        Rp=weights.dot(mean_price.T)
        SIGMAp=weights.dot(cov.dot(weights.T)) * len(r_data) 
        sha_r=(Rp-Rf)/np.sqrt(SIGMAp)
        if sha_r>best_shar:
            best_shar=sha_r
            best_weight=weights
            bset_rp=Rp
    final_couples.append(str(symbols))
    final_weights.append(best_weight)
    final_ratio.append(best_shar)
    final_rp.append(bset_rp)
    
result=pd.DataFrame()
result['portfolio']=final_couples
result['weights']=final_weights
result['sharp_ratio']=final_ratio
result['return']=final_rp
result=result.sort_values(by=['sharp_ratio'],ascending=False)
result=result.reset_index(drop=True)
result
```

## Sharp Ratio > 2, portfolio set 1
